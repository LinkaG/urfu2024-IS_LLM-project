```python
import os
import json
```

```python
# Путь к папке с данными
data_folder = 'collected_generated_text'
```

```python
# Проверяем, существует ли папка с данными
if os.path.exists(data_folder) and os.path.isdir(data_folder):
    # Получаем список файлов в папке
    files = os.listdir(data_folder)
    
    # Проходим по каждому файлу
    for file_name in files:
        print(f"Обрабртка файла {file_name}:")
        
        if file_name.endswith('.json'):  # Проверяем, что файл имеет расширение .json
            file_path = os.path.join(data_folder, file_name)
            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    try:
                        data = json.load(file)

                        total_rows = len(data)

                        correct_predictions_binary_classification = 0
                        correct_predictions_multiclass = 0
                        
                        for index, item in enumerate(data, 1):  # Проходимся по каждому элементу списка
                            if 'Expected CWE ID:' in item:expected_cwe_id = item['Expected CWE ID:']
                            if 'generated_text' in item: generated_text = item['generated_text']
                            if 'vulnerability: YES' in generated_text:
                                correct_predictions_binary_classification += 1
                                if any(f'vulnerability type: {expected_cwe_id}{punct}' in generated_text for punct in [',', '.', '!', '?', ':', ';', ' ', '\n']):
                                    correct_predictions_multiclass += 1
                                    
                        accuracy_binary_classification = (correct_predictions_binary_classification / total_rows) * 100
                        print(f"Точность результатов бинарной класификации: {accuracy_binary_classification:.2f}%/100%")
                        
                        accuracy_multiclass = (correct_predictions_multiclass / total_rows) * 100
                        print(f"Точность результатов многоклассовой класификации: {accuracy_multiclass:.2f}%/100%")

                        print(f"Обработано семплов: {len(data)}")
                    
                    except Exception as e:
                        print(f"Ошибка при обработке файла {file_name}: {e}")
            except Exception as e:
                print(f"Ошибка при открытии файла {file_name}: {e}")
        # break
else:
    print(f"Папка {data_folder} не существует или не является директорией")
```
